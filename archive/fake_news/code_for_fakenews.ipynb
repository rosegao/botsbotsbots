```python
from twitterscraper import query_tweets
import pandas as pd
from sklearn.model_selection import train_test_split
import datetime as dt
```

```python
query = query_tweets("Trump", limit=10000, begindate=dt.date(2018, 1, 1), enddate=dt.date(2018, 2, 11), lang="E")
```

```python
tweets_list = []
for tweet in query:
    tweets_list.append((tweet.fullname,
                        tweet.id,
                        tweet.likes,
                        tweet.replies,
                        tweet.retweets,
                        tweet.text,
                        tweet.timestamp,
                        tweet.user))
```

```python
tweet_columns = ['fullname',
                 'id',
                 'likes',
                 'replies',
                 'retweets',
                 'text',
                 'timestamp',
                 'user']
```

```python
df = pd.DataFrame(tweets_list, columns = tweet_columns)
```

```python
# Read in the fake news database
news_df = pd.read_csv("fakenews.csv", header=None, usecols=[0, 1, 2, 3])
```

```python
news_df.columns = ['site', 'category', 'category2', 'category3']
```

```python
# Filter the dataset for only definitively fake news sites - excludes 'bias', 'unreliable', 'political'
news_df = news_df[news_df.category.isin(['conspiracy', 'Conspiracy', 'fake',
       'fake news', 'hate', 'rumor', 'state'])]
```

```python
sites = list(news_df.site.unique())
```
